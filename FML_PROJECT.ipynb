# **Decision tree**

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

b = data.drop('satisfaction', axis=1).values
n = data['satisfaction']

# Split the data into training and testing sets
b_train, b_test, n_train, n_test = train_test_split(b, n, test_size=0.2, random_state=42)

# Create the DMatrix for XGBoost
dtrain = xgb.DMatrix(b_train, label=n_train)
dtest = xgb.DMatrix(b_test, label=n_test)

# Set the parameters for XGBoost
params = {
    'max_depth': 3,
    'eta': 0.1,
    'objective': 'binary:logistic',
    'eval_metric': 'logloss'
}

# Train the XGBoost model
num_rounds = 100
xgboost_model = xgb.train(params, dtrain, num_rounds)

# Make predictions on the testing data
n_pred = xgboost_model.predict(dtest)
n_pred_binary = [round(value) for value in n_pred]

# Calculate the accuracy of the model
accuracy = accuracy_score(n_test, n_pred_binary)
print("Accuracy:", accuracy)
```

```python
print(classification_report(n_test, n_pred_binary))
```
