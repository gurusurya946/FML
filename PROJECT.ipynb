changing the string data-types to numeric data-types

```python
data['satisfaction']=data['satisfaction'].replace('satisfied',1)
data['satisfaction']=data['satisfaction'].replace('dissatisfied',0)
print(data['satisfaction'])
```

```python
data['Gender']=data['Gender'].replace('Male',1)
data['Gender']=data['Gender'].replace('Female',0)
print(data['Gender'])
```

```python
data['Customer Type']=data['Customer Type'].replace('Loyal Customer',1)
data['Customer Type']=data['Customer Type'].replace('disloyal Customer',0)
print(data['Customer Type'])
```

```python
data['Class']=data['Class'].replace('Eco',1)
data['Class']=data['Class'].replace('Business',0)
data['Class']=data['Class'].replace('Eco Plus',2)
data['Class']
```

```python
data['Type of Travel']=data['Type of Travel'].replace('Personal Travel',1)
data['Type of Travel']=data['Type of Travel'].replace('Business travel',0)
data['Type of Travel']=data['Type of Travel'].replace('Personal Tr',2)
```

```python
#graph shows the output variable values
sns.distplot(data['satisfaction'], bins=30)
plt.show()
```

```python
#to know about the relationship between the attributes 
cr=data.corr()
cr
```

```python
#checking shape
data.shape
```

```python
#replacing the null values in attributes with the mean value of the respective attributes 
data['Arrival Delay in Minutes'].fillna(data['Arrival Delay in Minutes'].mean(),inplace=True)
data['Baggage handling'].fillna(data['Baggage handling'].mean(),inplace=True)
data['Checkin service'].fillna(data['Checkin service'].mean(),inplace=True)
data['Cleanliness'].fillna(data['Cleanliness'].mean(),inplace=True)
data['Online boarding'].fillna(data['Online boarding'].mean(),inplace=True)
data['Departure Delay in Minutes'].fillna(data['Departure Delay in Minutes'].mean(),inplace=True)
     
```

```python
# checking is all null values are cleared or not
data.isnull().sum()
```

```python
#CORRELATOION MATRIX TO UNDERSTAND THE REALTIONSHIP BETWEEN ATTRIBUTES IN BETTER WAY
plt.figure(figsize=(21, 21))
sns.heatmap(cr, annot=True)
plt.show()
     
```

```python
#dividing table attributes as input and output variables
X=data.drop('satisfaction',axis=1).values
y=data['satisfaction'].values
print(X)
```

```python
#spliting the data into training and testing 
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=42)
     
```

```python
colnames = list(data.columns)
colnames.remove('satisfaction')
```

```python
#applying logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
X_train.reshape(-1,1)
y_train.reshape(-1,1)
reg=LogisticRegression(C=0.5)
reg.fit(X_train,y_train)
```

```python
#to display coffecient
reg.coef_
```

```python
#to display intercept
reg.intercept_scaling
```

```python
#probability for X_train
reg.predict_proba(X_train)
```

```python
#for y_predict
y_predict=reg.predict(X_test)
print(y_predict)
```

```python
#to display score
reg.score(X_test, y_test)
```

```python
#to display confusion matrix
confusion_matrix(y_test,y_predict)
```

```python
#to display classification report
print(classification_report(y_test, y_predict))
```

applying random forest method to same dataset

```python
a=data.drop('satisfaction',axis=1).values
b=data['satisfaction'].values
print(a)
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.01, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(a_train, b_train)
```

```python
b_pred = rf.predict(a_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(b_test, b_pred))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
confusion_matrix(b_test, b_pred)
```

```python
print(classification_report(b_test, b_pred))
```
